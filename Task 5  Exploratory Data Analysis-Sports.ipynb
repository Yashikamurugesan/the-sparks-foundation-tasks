```python
four_data=complete_data[complete_data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by four','sum'),('fours','count')])
```

```python
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')


plt.xlabel('player name',fontsize=20)
plt.ylabel("No of hours",fontsize=20)
plt.title('Numbers of fours hit by players',fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

```python
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',
                                                                                          kind='bar',color='brown')

plt.xlabel('season',fontsize=20)
plt.ylabel("No of hours",fontsize=20)
plt.title('Numbers of fours hit in each season',fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

```python
six_data=complete_data[complete_data["batsman_runs"]==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('season','count')])
```

```python
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='purple')


plt.xlabel('player name',fontsize=20)
plt.ylabel("No of hours",fontsize=20)
plt.title('Numbers of six hit by players',fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

```python
ax=six_data.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color='blue')

plt.xlabel('season',fontsize=20)
plt.ylabel("No of hours",fontsize=20)
plt.title('Numbers of fours hit in each season',fontsize=20)
plt.xticks(rotation=50)
plt.show()
```
